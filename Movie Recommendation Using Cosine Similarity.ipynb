```python
import pandas as pd
from functools import reduce
import ast
```

```python
credits = pd.read_csv('credits.csv')
keywords = pd.read_csv('keywords.csv')
links =pd.read_csv('links.csv')
movies = pd.read_csv('movies_metadata.csv')
rating =pd.read_csv('ratings.csv')
rating_small =pd.read_csv('ratings_small.csv')
```

```python
rating_small.rename(columns={'movieId': 'id'}, inplace=True)
links.rename(columns={'movieId': 'id'}, inplace=True)
```

```python
dfs = [movies, credits, keywords, links, rating_small]

dfs = [movies, credits, keywords, links, rating_small]

for df in dfs:
    df['id'] = df['id'].astype(str)  


movies_merged = reduce(lambda left, right: pd.merge(left, right, on='id', how='inner'), dfs)

movies_merged.head()
```

```python
movies = movies_merged[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python

movies['keywords']
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)[:3]])
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags']
```

```python
movies = movies[['id', 'title', 'overview', 'tags']]
```

```python
movies = movies.drop_duplicates()
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
movies.head()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```
